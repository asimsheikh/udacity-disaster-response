```python
import pandas as pd
```

```python
messages_df = pd.read_csv('./data/disaster_messages.csv')
```

```python
messages_df.head()
```

```python
categories_df = pd.read_csv('./data/disaster_categories.csv')
```

```python
categories_df.head()
```

```python
df = pd.merge(messages_df, categories_df, on='id')
```

```python
categories_df = categories_df['categories'].str.split(';', expand=True)
```

```python
categories_df.head()
```

```python
categories_df.iloc[0].apply(lambda text: text[:-2])
```

```python
categories_df.columns = categories_df.iloc[0].apply(lambda text: text[:-2])
```

```python
categories_df.head()
```

```python
categories_df[categories_df.columns[0]].apply(lambda text: text.split('-')[-1])
```

```python
for column in categories_df:
    categories_df[column] = categories_df[column].apply(lambda text: text.split('-')[-1])
    categories_df[column] = categories_df[column].astype(int)
```

```python
categories_df.info()
```

```python
names = pd.DataFrame( {'names': ['Asim', 'Faizan', 'Khurrum']})
ages = pd.DataFrame( {'ages': [42, 34, 36]})
pd.concat([names, ages], axis='columns')
```

```python
# pd.concat([df, categories_df], axis='columns')
df.drop('categories', axis='columns', inplace=True)
```

```python
df = pd.concat([df, categories_df], axis='columns')
```

```python
df.duplicated().sum()
```

```python
filt = df.duplicated()
df = df[~filt]
```
